### Cleaning 

These are some of the quality and tidiness issues we encountered after thoroughly accessing the datasets

### Quality issues
1. The in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id  retweeted_status_user_id  and retweeted_status_timestamp columns in the twitter-archive-enhanced dataset have more than 90% of its data been null values
2. Source of tweet embedded in the href tag in the twitter-archive-enhanced dataset.

3. Inconsistent rating_numerator amd rating_denominator values compared to what is in the text column.

4. Incorrect datatype for the ID columns in all 3 dataset 

5. Incorrect datatype for the new rating_numerator and rating_denominator  values 

6. Incorrect datatype for lang column in enhance_df2 dataframe and timestamp in enhance_df 

7. All three predictions in the image_prediction_df dataframe been false

8. Actual rating of each dog

### Tidiness issues
9. Dog stages in 4 columns instead of 1

10. All dataframes form one observation

#### Solving Issue 1

I dropped the in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id, retweeted_status_user_id and retweeted_status_timestamp columns because their entries were not up to 10% of the total entries needed. The new dataframe will have now have the columns displayed below;

#### Solving Issue 2

The source of the tweet was in an href tag. I stripped the tags off with a regular expression leaving the actual content.

#### Solving Issue 3

The actual values for the rating_numerator and rating_denominator can be found in the text column. Since some of the original values given for these columns were said to be incorrect, I had to extract the new values from the text column using a regular expression. #### Solving Issue 4

Since we are not going to perform any mathematical operations on the id columns, it is best we convert its datatype to object. This I did and also change the column name 'id' in the twitter-archive_2 dataset to tweet_id to make it consistnt with the other two dataframes

#### Solving Issue 5

Our columns rating_numerator_new and rating_denominator_new came with the object datatype. For the sake of further analysis on this column, we change the datatype float.

#### Solving Issue 6

Another datatype isssue is the lang and timestamp columns in the twitter_archive_2 and twitter_archive dataset respectively. I changed the lang to category datatype and timestamp to dtae_time.

#### Solving Issue 7

If all the three algorithms return false, it means there is a high chance of the image tweeted not been a dog or even having a dog in it. These rows will not be very beneficial in our analysis so I dropped them from the image_prediction dataset

#### Solving Issue 8

For the sake of analysis, I ought to calculate the actual rating of each dog. What we have is fractional parts with each part in a seperate column. I divided the rating_numerator_new with rating_denominator_new to get the actual rating of each dog and stored the results in a new column. The new column, aka rating was then appended to the twitter_archive dataset.

#### Solving Issue 9

The doggo,floofer,pupper,puppo columns in the enhanced_df dataset all depicts the stage of growth of the dog. The 4 dataset should be actually 1 column with its name as dog_stage. The datatype was then changed to category data type. Below is the display of the columns in the twitter_archive dataset after the dog_stage column is created and the other 4 is deleted.
